```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
theMovies=pd.read_csv('../DataSources/theMovieFeatures.csv')
theMovies_qual=pd.read_csv('../DataSources/theMovieFeatures_qual.csv')
theMovies.columns
```

```python
total_winners=theMovies[theMovies['Best Picture']==1].shape[0]
total_losers=theMovies[theMovies['Best Picture']==0].shape[0]

winners=theMovies[theMovies['Best Picture']==1]
losers=theMovies[theMovies['Best Picture']==0]

winners_BestActor=winners[winners['Best Actor']==1].shape[0]
winners_BestActress=winners[winners['Best Actress']==1].shape[0]
winners_BestSuppActor=winners[winners['Best Supporting Actor']==1].shape[0]
winners_BestSuppActress=winners[winners['Best Supporting Actress']==1].shape[0]
winners_Directing=winners[winners['Best Director']==1].shape[0]

per_BestActor=winners_BestActor/(total_winners)*100
per_BestActress=winners_BestActress/(total_winners)*100
per_BestSuppActor=winners_BestSuppActor/(total_winners)*100
per_BestSuppActress=winners_BestSuppActress/(total_winners)*100
per_BestDir=winners_Directing/(total_winners)*100

Awards= [per_BestDir, per_BestActor, per_BestSuppActor, per_BestSuppActress, per_BestActress]
Award_Labels = ['Best Director', 'Best Actor', 'Best Supporting Actor', 'Best Supporting Actress', 'Best Actress']

plt.figure(figsize=(15,5))
AwardsPlot = plt.bar(Award_Labels, Awards, color='gold', edgecolor='black', linewidth=2)

plt.xlabel('Awards', size=15)
plt.xticks(size=15, rotation=45)
plt.ylabel('% of Best Picture Winners', size=15)
plt.yticks(size=15)
plt.title('% of Best Pictures Winners that Won Another Award', size=20)
plt.savefig('../PlotsAndImages/Awards.png')
plt.show()
```

```python
Yearly_Trends_nominees=theMovies.groupby('Year', as_index=False).agg({'Title':np.size,
                                                                      'Rated_APPROVED': np.sum,
                                                                     'Rated_G': np.sum,
                                                                     'Rated_PG': np.sum,
                                                                     'Rated_PG-13':np.sum,
                                                                     'Rated_R': np.sum,
                                                                     'Rated_UNRATED':np.sum}
                                                                    ).rename(columns={'Title':'Total Nominees'})

Yearly_Trends_nominees['Total Ratings']=Yearly_Trends_nominees['Rated_APPROVED']+Yearly_Trends_nominees['Rated_G']+Yearly_Trends_nominees['Rated_PG']+Yearly_Trends_nominees['Rated_PG-13']+Yearly_Trends_nominees['Rated_R']+Yearly_Trends_nominees['Rated_UNRATED']
Yearly_Trends_nominees['No Rating Available']=Yearly_Trends_nominees['Total Nominees']-Yearly_Trends_nominees['Total Ratings']

x_axis=Yearly_Trends_nominees['Year']
y_axis1=Yearly_Trends_nominees['Rated_APPROVED']
y_axis2=Yearly_Trends_nominees['Rated_G']
y_axis3=Yearly_Trends_nominees['Rated_PG']
y_axis4=Yearly_Trends_nominees['Rated_UNRATED']
y_axis5=Yearly_Trends_nominees['Rated_PG-13']
y_axis6=Yearly_Trends_nominees['Rated_R']
y_axis7=Yearly_Trends_nominees['No Rating Available']

barWidth = 1
plt.figure(figsize=(17,8))
plt.bar(x_axis, y_axis1, edgecolor='white', color='palegreen', width=barWidth, label='Approved')
plt.bar(x_axis, y_axis2, bottom=y_axis1, edgecolor='white', color='lawngreen', width=barWidth, label='G')
plt.bar(x_axis, y_axis3, bottom=y_axis1+y_axis2, edgecolor='white', color='lime', width=barWidth, label='PG')
plt.bar(x_axis, y_axis4, bottom=y_axis1+y_axis2+y_axis3, edgecolor='white', color='limegreen', width=barWidth, label='Unrated')
plt.bar(x_axis, y_axis5, bottom=y_axis1+y_axis2+y_axis3+y_axis4, edgecolor='white', color='orange', width=barWidth, label='PG-13')
plt.bar(x_axis, y_axis6, bottom=y_axis1+y_axis2+y_axis3+y_axis4+y_axis5, edgecolor='white', color='orangered', width=barWidth, label='R')
plt.bar(x_axis, y_axis7, bottom=y_axis1+y_axis2+y_axis3+y_axis4+y_axis5+y_axis6, edgecolor='white', color='tan', width=barWidth, label='Rating Not Available')

plt.xlim(1940, 2020)
plt.ylim(0,11)
plt.ylabel('Number of Nominees', size=15)
plt.xticks(size=15)
plt.yticks(size=15)
plt.legend(loc="upper left", fontsize="large", fancybox=True)
plt.title('Number of Nominees by Rating', size=20)
plt.savefig('../PlotsAndImages/NumRatings.png')
plt.show()
```

```python
#Analyze Movie Genres
#groupby Best Picture Win/Loss and Genre
grouped=theMovies_qual.groupby(['Best Picture', 'Genre'], as_index=False).size()
#unstack to get Wins and Losses side by side
genre_df=grouped.unstack(0, fill_value=0).rename(columns={0:'Losses', 1:'Wins'})
#Calculate % Wins
genre_df['% Wins']=genre_df['Wins']/(genre_df['Wins']+genre_df['Losses'])*100
genre_df = genre_df.sort_values('% Wins', ascending=False)


plt.bar(genre_df.index,genre_df["% Wins"],color='gold',edgecolor='black',linewidth=2)
plt.xticks(genre_df.index,rotation=90,size=15)
plt.yticks(size=15)
plt.title("% of Best Picture Wins by Genre Classifcation",size=15)
plt.ylabel("% Wins",size=15)
# plt.figure(figsize=(25,25))
plt.savefig("../PlotsAndImages/BestPicturebyGenre.png")
plt.show()
```

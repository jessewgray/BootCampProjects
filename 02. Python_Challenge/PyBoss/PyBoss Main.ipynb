```python
import string
import csv
import os

us_state_abbrev = {
        'Alabama': 'AL',
        'Alaska': 'AK',
        'Arizona': 'AZ',
        'Arkansas': 'AR',
        'California': 'CA',
        'Colorado': 'CO',
        'Connecticut': 'CT',
        'Delaware': 'DE',
        'Florida': 'FL',
        'Georgia': 'GA',
        'Hawaii': 'HI',
        'Idaho': 'ID',
        'Illinois': 'IL',
        'Indiana': 'IN',
        'Iowa': 'IA',
        'Kansas': 'KS',
        'Kentucky': 'KY',
        'Louisiana': 'LA',
        'Maine': 'ME',
        'Maryland': 'MD',
        'Massachusetts': 'MA',
        'Michigan': 'MI',
        'Minnesota': 'MN',
        'Mississippi': 'MS',
        'Missouri': 'MO',
        'Montana': 'MT',
        'Nebraska': 'NE',
        'Nevada': 'NV',
        'New Hampshire': 'NH',
        'New Jersey': 'NJ',
        'New Mexico': 'NM',
        'New York': 'NY',
        'North Carolina': 'NC',
        'North Dakota': 'ND',
        'Ohio': 'OH',
        'Oklahoma': 'OK',
        'Oregon': 'OR',
        'Pennsylvania': 'PA',
        'Rhode Island': 'RI',
        'South Carolina': 'SC',
        'South Dakota': 'SD',
        'Tennessee': 'TN',
        'Texas': 'TX',
        'Utah': 'UT',
        'Vermont': 'VT',
        'Virginia': 'VA',
        'Washington': 'WA',
        'West Virginia': 'WV',
        'Wisconsin': 'WI',
        'Wyoming': 'WY',
        }
```

```python
#funciton oto anonymize the first 5 digits of SSN
def SSN_replace (SSN):
    SSN_split=SSN.split("-")
    for digits in SSN_split:
        SSN_split[0]="xxx"
        SSN_split[1]="xx"
    anon_SSN='-'.join(SSN_split)
    return anon_SSN

#function oto change date format from yyyy-mm-dd to mm/dd/yyyy
def date_format (yyyymmdd):
    date=yyyymmdd.split("-")
    reorder=[1,2,0]
    date=[date[i] for i in reorder]
    mmddyyyy='/'.join(date)
    return mmddyyyy

#function to replace state name with state abbreviation
def state_abbrev(state):
    for state_name in us_state_abbrev:
        if state==state_name:
            state_abbrev=state_name[0]+state_name[1]
    return state_abbrev
```

```python
csvpath=os.path.join("employee_data.csv")

with open (csvpath, 'r', newline='') as csvfile:
    employee_data=csv.reader(csvfile, delimiter=",")
    next(employee_data, None)

    #create empty lists to populate munged data
    SSNs=[]
    DOBs=[]
    states=[]
    first_names=[]
    last_names=[]
    ids=[]
    
    for item in employee_data:
        
        #cahnge all SSN numbers in employee_data to anonymized version and populate in new list
        new_SSN=SSN_replace(item[3])
        SSNs.append(new_SSN)
        
        #change all date of births in employee_data to correct date format and populate in new list
        new_DOB=date_format(item[2])
        DOBs.append(new_DOB)

        #change all state names in employee_data to state abbreviations and populate in new list 
        new_st=state_abbrev(item[4])
        states.append(new_st)
        
        #split first and last names in employee_data and populate each in their own new list
        new_name=item[1].split(' ')
        first_names.append(new_name[0])
        last_names.append(new_name[1])
        
        #Put all employee IDs from employee_data in their own list
        new_id=item[0]
        ids.append(new_id)
```

```python
#zip all new lists created above into single list 
new_employee_data=zip(ids, first_names, last_names, DOBs, SSNs, states)

#output new employee data to csv file
output_file = os.path.join("new_employee_data.csv")
with open(output_file, "w", newline="") as datafile:
    writer = csv.writer(datafile)
    writer.writerow(["Employee ID", "First Name", "Last Name", "DOB",
                         "Anon SSN", "State"])
    writer.writerows(new_employee_data)
```
